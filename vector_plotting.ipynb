```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.quiver(0,0,3,4)
plt.show()
```

```
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1)
plt.show()
```

```
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,4,scale_units='xy',angles='xy',scale=1,color='g')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
def plot_vector(vecs):
  color=['r','b','g','y']
  i=0;
  for vec in vecs:
    plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units='xy',angles='xy',scale=1,color=color[i%len(color)])
    i+= 1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
plot_vector([ (0,0,3,4) , (0,0,-3,4),(0,0,-3,2) ])
```

## **addition and subtraction of a vector**

```
vecs=[ (0,0,3,4) , (0,0,-3,4),(0,0,-3,2)]
```

```
plot_vector([vecs[0],vecs[1]])
```

```
vecs[0]+vecs[1] #here tuples are concatenate so we can make them as numpy array then do the opration
```

```
vecs=[ np.asarray( [0,0,3,4] ) , np.asarray( [0,0,-3,4] ),np.asarray( [0,0,-3,2])]
```

```
plot_vector([vecs[0],vecs[2]])
```

```
vecs[0]+vecs[2] 
```

```
plot_vector([ vecs[0], vecs[1],vecs[0]+vecs[2] ])
```

```
plot_vector([ vecs[0], vecs[1],vecs[0]-vecs[1] ])
```

```
plot_vector([ vecs[0], -vecs[1],vecs[0]+ (-vecs[1]) ])
```

## Dot product

```
a=np.asarray([3,4])
b=np.asarray([4,-1])
```

```
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

# projection of a vector

```
a_b=np.dot(a,b)/np.linalg.norm(b)
print(a_b)
```

```
vect_a_b =((a_b)/np.linalg.norm(b))*b
print(vect_a_b)
```
